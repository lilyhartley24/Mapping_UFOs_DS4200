Static: bar graph of the distribution of the shapes of the UFOs

```python
# imports
import altair as alt
import pandas as pd
import matplotlib.pyplot as plt

# read data
data = pd.read_csv("archive/scrubbed.csv")

# transform datetime column to datetime (currently stored as a string)
data["datetime"] = pd.to_datetime(data["datetime"], errors="coerce")

# filter to only US data
project_df = data[data["country"] == "us"]

# filter data from 2005 onwards (window of 10 years of data)
recent_df = project_df[project_df["datetime"].dt.year >= 2005]

recent_df
```

```python
# bar graph of distribution of shapes of UFOs

shape_counts = recent_df["shape"].value_counts().head(10)

plt.figure(figsize=(10,6))
plt.bar(shape_counts.index, shape_counts.values, color="darkorange")

plt.xticks(rotation=45)
plt.xlabel("Shape")
plt.ylabel("Count")
plt.title("Top 10 UFO Shapes (2005â€“2015)")
plt.tight_layout()
plt.savefig("Top_10_Shapes_Bar.png")
plt.show()
```
